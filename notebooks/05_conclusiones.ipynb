# Conclusiónes del Proyecto

## 1. Introducción

El presente Proyecto Integrador permitió recorrer de manera completa el ciclo de trabajo propuesto en la asignatura Minería de Datos 1, desde la inspección inicial del conjunto de datos hasta la comunicación de resultados mediante una aplicación interactiva. A lo largo del proyecto se desarrolló un proceso reproducible de extracción, limpieza, transformación y análisis de información, aplicando técnicas de calidad de datos, análisis exploratorio y reducción de dimensionalidad.

Más allá de los resultados obtenidos, el trabajo permitió demostrar cómo un proceso metodológicamente estructurado transforma un conjunto de datos con inconsistencias en una fuente confiable de información para la toma de decisiones.

## 2. Objetivos y Preguntas Iniciales:
- Construir un flujo de trabajo reproducible que garantice la confiabilidad de los datos.
    - ¿Es posible limpiar y estandarizar el dataset reduciendo inconsistencias sin perder representatividad?

- Identificar patrones de consumo y perfiles de uso en la plataforma.
    - ¿Cómo influyen la edad, el tipo de suscripción y la región en el comportamiento y uso del soporte por parte del usuario?

- Evaluar si es viable reducir las dimensiones del dataset analizando las relaciones de sus variables numéricas.
    - ¿Existen variables latentes o componentes principales que permitan resumir el comportamiento del usuario sin pérdida crítica de información?

## 2. Principales Hallazgos
### Hallazgo 1: Calidad y preparación de los datos

El proceso ETL permitió depurar exitosamente el conjunto de datos original, eliminando registros duplicados, corrigiendo inconsistencias e imputando valores faltantes según el mecanismo de ausencia identificado.

**Evidencia:** Eliminación de registros duplicados e inconsistentes. Corrección de edades imposibles y fechas inválidas. Tratamiento diferenciado de datos MCAR y MAR. Retención final del 93.54 % de los registros originales.

**Interpretación:** La elevada tasa de retención demuestra que fue posible mejorar significativamente la calidad del dataset sin comprometer su representatividad. Además, las decisiones de limpieza se apoyaron en criterios estadísticos y metodológicos, evitando eliminar información de manera arbitraria.

**Conclusión:** El pipeline ETL construido constituye una base sólida y reproducible para realizar análisis posteriores con datos confiables.

### Hallazgo 2: Comportamiento de los usuarios

El análisis exploratorio permitió identificar patrones relevantes en el comportamiento de los usuarios de la plataforma.

**Evidencia:** La mayor concentración de usuarios se encuentra entre los 25 y 45 años. Los clientes Premium registran el mayor consumo mensual. El comportamiento del soporte técnico presenta diferencias entre países.

**Interpretación:** Estos resultados muestran que el nivel de suscripción está asociado con una mayor intensidad de uso, mientras que las diferencias observadas entre mercados sugieren que la experiencia del usuario puede estar influenciada por factores regionales, operativos o comerciales.

**Conclusión:** El análisis exploratorio permitió transformar datos descriptivos en información útil para comprender mejor el comportamiento de los usuarios y detectar oportunidades de mejora en la operación del servicio.

### Hallazgo 3: Reducción de dimensionalidad mediante PCA

El análisis de Componentes Principales permitió estudiar la estructura interna de las variables numéricas luego de aplicar un escalamiento estándar.

**Evidencia:** Los tres componentes principales explican aproximadamente un tercio de la varianza cada una. Ninguna componente concentra suficiente información como para reemplazar a las restantes. Las cargas factoriales permitieron identificar patrones diferenciados de comportamiento.

**Interpretación:** Aunque el PCA no logró reducir la dimensionalidad sin pérdida considerable de información, sí permitió sintetizar las relaciones existentes entre las variables y describir tres perfiles claramente diferenciados:
* PC1: actividad general del usuario.
* PC2: usuarios jóvenes con mayor interacción con soporte.
* PC3: usuarios con alto consumo y baja interacción con soporte.

**Conclusión:** En este caso, el principal aporte del PCA no fue la reducción de variables, sino la interpretación de la estructura de los datos y la identificación de perfiles de comportamiento.

## 3. Limitaciones del Estudio

Como todo proyecto analítico, este trabajo presenta algunas limitaciones que deben considerarse al interpretar los resultados.

El análisis se realizó sobre un conjunto limitado de variables disponibles en el dataset original.
No se dispone de información histórica suficiente para construir modelos predictivos.
Algunas variables relevantes para el negocio, como método de pago, permanencia del cliente o dispositivo utilizado, no estaban disponibles.
El estudio se centró exclusivamente en técnicas descriptivas y de reducción de dimensionalidad, sin incorporar modelos supervisados.

Estas limitaciones no afectan la validez del trabajo realizado, pero representan oportunidades claras para ampliar el alcance del proyecto en futuras etapas.

## 4. Próximos Pasos

A partir de los resultados obtenidos, se identifican diversas oportunidades para continuar desarrollando este proyecto.

* Aplicar técnicas de Clustering utilizando las componentes principales para segmentar usuarios con comportamientos similares.
* Incorporar información histórica para desarrollar modelos predictivos de abandono de clientes (Churn).
* Automatizar completamente el pipeline ETL para procesar nuevos archivos y generar alertas cuando la calidad de los datos disminuya.
* Enriquecer el conjunto de datos incorporando variables como método de pago, dispositivo de reproducción y antigüedad del cliente.

## 5. Conclusión Final

El proyecto permitió construir un proceso analítico completo, reproducible y fundamentado, integrando todas las etapas abordadas en la materia: calidad de datos, preparación, análisis exploratorio, reducción de dimensionalidad y comunicación de resultados.

Desde una perspectiva metodológica, se demostró la importancia de realizar un proceso de limpieza riguroso antes de aplicar cualquier técnica analítica. Desde una perspectiva de negocio, el estudio permitió identificar diferencias significativas en los patrones de consumo y en la interacción con el soporte técnico entre distintos segmentos de usuarios.

Si bien el análisis de Componentes Principales mostró que era necesario conservar las tres componentes para mantener prácticamente toda la información del conjunto de datos, también permitió descubrir perfiles de comportamiento que aportan una comprensión más profunda de los usuarios de la plataforma.

En conjunto, los resultados obtenidos constituyen una base sólida para futuras aplicaciones de segmentación, analítica predictiva y apoyo a la toma de decisiones, demostrando el valor de combinar técnicas de preparación de datos, análisis exploratorio y aprendizaje no supervisado dentro de un mismo flujo de trabajo reproducible.
